## Random Walks

```python
np.random.standard_normal(10)  # legacy way
```

```python
np.random.random(10)  # legacy way
```

```python
from numpy.random import default_rng
```

```python
rng = default_rng()
```

```python
# rng = default_rng(seed=100)  # with fixed seed (for reproducibility)
```

```python
rng.standard_normal(20)
```

```python
%time rn = rng.standard_normal(10000)
```

```python
rn.mean()
```

```python
rn.std()
```

```python
rn_ = (rn - rn.mean()) / rn.std()
```

```python
rn_.mean().round(10)
```

```python
rn_.std()
```

```python
rw = rng.standard_normal((500, 5))
```

```python
rw[:6]
```

```python
rw[0] = 0.0
```

```python
rw[:6]
```

```python
rw = rw.cumsum(axis=0)  # random walk data
```

## Reading Financial Data

```python
url = 'https://certificate.tpq.io/mlfin.csv'
```

```python
raw = pd.read_csv(url, index_col=0, parse_dates=True)
```

```python
raw.head()
```

```python
raw.tail()
```

```python
raw.dropna().tail()
```

```python
raw['AAPL.O'].dropna().plot();
```

```python
# raw.plot?
```

```python
raw['AAPL.O'].iloc[-100:].plot(style='r.');
```

```python
raw[['.SPX', '.VIX']].dropna().plot(secondary_y='.VIX');
```

```python
raw[['AAPL.O', 'AMZN.O']].dropna().plot();
```

```python
(raw[['.SPX', 'AAPL.O']] / raw[['.SPX', 'AAPL.O']].iloc[0]).dropna().plot();
```

```python
(raw[['AMZN.O', 'AAPL.O']] / raw[['AMZN.O', 'AAPL.O']].iloc[0]).dropna().plot();
```

```python
del raw['BTC=']
```

```python
raw.dropna(inplace=True)
```

```python
rets = np.log(raw / raw.shift(1))
```

```python
# raw.shift(1)
```

```python
rets.head()
```

```python
rets.mean()
```

```python
rets.mean().apply(np.exp) - 1  # daily net returns
```

```python
np.exp(rets.mean()) - 1  # daily net returns
```

```python
(rets.mean().apply(np.exp) - 1) * 252  # annualized
```

## Excursion

Simple vs. Log Returns

```python
S0 = 100
```

```python
S1 = 110
```

```python
S1 / S0  # gross return
```

```python
S1 / S0 - 1 # net return
```

```python
lr = math.log(S1 / S0)  # net log return
lr
```

```python
math.exp(lr)  # gross return
```

```python
math.exp(lr) - 1  # net return
```
